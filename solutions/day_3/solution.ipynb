```python
from string import digits
from string import ascii_letters 
```

```python
INPUT_FILE = 'input.txt'

with open(INPUT_FILE, 'r') as f:
    lines = [line.strip('\n') for line in f.readlines()]

# lines
len(lines)
```

```python
special_chars = []

def find_special_chars():
    for line in lines:
        for char in line:
            if char != '.' and char not in digits:
                special_chars.append(char)


find_special_chars()

special_chars = set(special_chars)
special_chars
```

```python
def get_digit_coords():
    digit_coords = []
    for i, line in enumerate(lines):
        for j, char in enumerate(line):
            if char in digits:
                digit_coords.append([i, j])
    return digit_coords
```

```python
def get_special_char_coords():
    special_char_coords = []
    for i, line in enumerate(lines):
        for j, char in enumerate(line):
            if char in special_chars:
                special_char_coords.append([i, j])
    return special_char_coords
```

```python
import math as m
import re
from string import digits

board = list(open('input.txt'))
chars = {(r, c): [] for r in range(140) for c in range(140) if board[r][c] not in '0123456789.'}

for r, row in enumerate(board):
    for n in re.finditer(r'\d+', row):
        edge = {(r, c) for r in (r-1, r, r+1)
                       for c in range(n.start()-1, n.end()+1)}
        for o in edge & chars.keys():
            chars[o].append(int(n.group()))

print(sum(sum(p)    for p in chars.values()),
      sum(m.prod(p) for p in chars.values() if len(p)==2))
```
